```python
import pandas as pd
import json
```

```python
with open('enriched_posts.json', 'r') as file:
    data = json.load(file)
```

```python
df = pd.DataFrame(data)
print(df.head())
```

```python
import pandas as pd
import nltk
import spacy
from collections import Counter
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
df.head()
```

```python
texts = df['text'].astype(str).tolist()
texts[:5]
```

```python
from nltk.tokenize import word_tokenize

tokenized_texts = [word_tokenize(text) for text in texts]

for i in range(3):
    print("Tokens:", tokenized_texts[i])
```

```python
nltk_pos_tags = []

for tokens in tokenized_texts:
    tags = nltk.pos_tag(tokens)
    nltk_pos_tags.append(tags)

for i in range(3):
    print("NLTK POS Tags:", nltk_pos_tags[i])
```

```python
nlp = spacy.load("en_core_web_sm")

spacy_pos_tags = []

for text in texts:
    # Clean up potential malformed unicode characters
    cleaned_text = text.encode('utf-8', errors='ignore').decode('utf-8')
    doc = nlp(cleaned_text)
    tags = [(token.text, token.pos_) for token in doc]
    spacy_pos_tags.append(tags)

for i in range(3):
    print("spaCy POS Tags:", spacy_pos_tags[i])
```

```python
comparison_df = pd.DataFrame({
    "Text": texts[:5],
    "NLTK_POS": nltk_pos_tags[:5],
    "spaCy_POS": spacy_pos_tags[:5]
})

comparison_df
```

```python
nouns = []

for doc in spacy_pos_tags:
    for word, pos in doc:
        if pos in ["NOUN", "PROPN"]:
            nouns.append(word.lower())

noun_freq = Counter(nouns)
print("Noun Frequency:", noun_freq)
```

```python
verbs = []

for doc in spacy_pos_tags:
    for word, pos in doc:
        if pos == "VERB":
            verbs.append(word.lower())

verb_freq = Counter(verbs)
print("Verb Frequency:", verb_freq)
```

```python
noun_df = pd.DataFrame(noun_freq.items(), columns=["Noun", "Frequency"])
verb_df = pd.DataFrame(verb_freq.items(), columns=["Verb", "Frequency"])

noun_df.head(), verb_df.head()
```
