```python
medical="""The patient is a 54-year-old male with a history of hypertension and diabetes mellitus who presented with complaints of chest pain radiating to the left arm. The pain started two hours prior to admission and was associated with shortness of breath and sweating. On examination, the patient was alert and oriented. Blood pressure was 150/90 mmHg, pulse rate was 88 beats per minute. Cardiac examination revealed normal heart sounds with no murmurs. ECG showed ST-segment elevation in the anterior leads. The patient was diagnosed with acute myocardial infarction and was started on aspirin, nitroglycerin, and heparin therapy."""
```

```python
print(medical)
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
word_tokenize(medical)
```

```python
from nltk.tokenize import sent_tokenize
sent_tokenize(medical)
```

```python

nltk.download("stopwords")
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
words_in_quote = word_tokenize(medical)
words_in_quote
```

```python
stop_words = set(stopwords.words("english"))
filtered_list = []
for word in words_in_quote:
  if word.casefold() not in stop_words:
    filtered_list.append(word)
filtered_list
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
stemmer = PorterStemmer()
words = word_tokenize(medical)
stemmed_words = [stemmer.stem(word) for word in words]
stemmed_words
```

```python
from nltk.stem import SnowballStemmer
snowball = SnowballStemmer(language='english')
words = word_tokenize(medical)
for word in words:
    print(word,"--->",snowball.stem(word))
```

```python
from nltk import LancasterStemmer
Lanc = LancasterStemmer()
words = word_tokenize(medical)
for word in words:
    print(word,"--->",Lanc.stem(word))
```

```python
from nltk.stem import RegexpStemmer
regexp = RegexpStemmer('''ing
|e''', min=4)
words = word_tokenize(medical)
for word in words:
    print(word,"--->",regexp.stem(word))
```

```python
nltk.download('omw-1.4')
nltk.download('wordnet') # Added to download the wordnet corpus
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
words = word_tokenize(medical)
for word in words:
    print(word,"--->",lemmatizer.lemmatize(word))
```

```python
lemmatizer.lemmatize("worst")
```

```python
lemmatizer.lemmatize("worst", pos="a")
```
